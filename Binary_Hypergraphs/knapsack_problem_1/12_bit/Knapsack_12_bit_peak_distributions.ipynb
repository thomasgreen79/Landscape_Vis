```r
library(tidyverse)
```

```r
raw_data_25 = read.csv("../data/knapsack_problem_12_bit_25_wt_data.csv")
raw_data_72 = read.csv("../data/knapsack_problem_12_bit_72_wt_data.csv")
raw_data_100 = read.csv("../data/knapsack_problem_12_bit_100_wt_data.csv")
raw_data_25_search_1 = read.csv("../data/knapsack_problem_12_bit_25_wt_actual_search_1_data.csv")
raw_data_25_all_searches = read.csv("../data/knapsack_problem_12_bit_25_wt_all_searches_data.csv")
```

```r
raw_data_25$eigenbin <- cut(raw_data_25$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
raw_data_72$eigenbin <- cut(raw_data_72$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
raw_data_100$eigenbin <- cut(raw_data_100$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
raw_data_25_search_1$eigenbin <- cut(raw_data_25_search_1$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
raw_data_25_all_searches$eigenbin <- cut(raw_data_25_all_searches$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
```

```r
raw_data_25 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
raw_data_72 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point() 
```

```r
raw_data_100 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
raw_data_25_all_searches %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
peak_data_25 = raw_data_25 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
peak_data_72 = raw_data_72 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
peak_data_100 = raw_data_100 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
peak_data_25_search_1 = raw_data_25_search_1 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
peak_data_25_all_searches = raw_data_25_all_searches %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
```
