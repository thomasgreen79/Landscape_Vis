```r
library(tidyverse)
```

```r
raw_data_0 = read.csv("../data/nk_problem_12_bit_0_kval_data.csv")
raw_data_3 = read.csv("../data/nk_problem_12_bit_3_kval_data.csv")
raw_data_7 = read.csv("../data/nk_problem_12_bit_7_kval_data.csv")
raw_data_11 = read.csv("../data/nk_problem_12_bit_11_kval_data.csv")
```

```r
raw_data_0$eigenbin <- cut(raw_data_0$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
raw_data_3$eigenbin <- cut(raw_data_3$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
raw_data_7$eigenbin <- cut(raw_data_7$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
raw_data_11$eigenbin <- cut(raw_data_11$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
```

```r
raw_data_0 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
raw_data_3 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
raw_data_7 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
raw_data_11 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
peak_data_0 = raw_data_0 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
peak_data_3 = raw_data_3 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
peak_data_7 = raw_data_7 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
peak_data_11 = raw_data_11 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
```

```r
peak_data_0 %>%
  ggplot(aes(x=eigenbin, y=peak_counts)) +
  geom_point()
```

```r
peak_data_3 %>%
  ggplot(aes(x=eigenbin, y=peak_counts)) +
  geom_point()
```

```r
peak_data_7 %>%
  ggplot(aes(x=eigenbin, y=peak_counts)) +
  geom_point()
```

```r
peak_data_11 %>%
  ggplot(aes(x=eigenbin, y=peak_counts)) +
  geom_point()
```

```r
peak_data_0 %>%
  ggplot(aes(x=eigenbin, y=percent)) +
  geom_point()
```

```r
peak_data_3 %>%
  ggplot(aes(x=eigenbin, y=percent)) +
  geom_point()
```

```r
peak_data_7 %>%
  ggplot(aes(x=eigenbin, y=percent)) +
  geom_point()
```

```r
peak_data_11 %>%
  ggplot(aes(x=eigenbin, y=percent)) +
  geom_point()
```
