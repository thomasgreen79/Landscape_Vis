```r
library(tidyverse)
```

```r
raw_data_30 = read.csv("../data/knapsack_problem_14_bit_30_wt_data.csv")
raw_data_77 = read.csv("../data/knapsack_problem_14_bit_77_wt_data.csv")
raw_data_90 = read.csv("../data/knapsack_problem_14_bit_90_wt_data.csv")
raw_data_120 = read.csv("../data/knapsack_problem_14_bit_120_wt_data.csv")
```

```r
raw_data_30$eigenbin <- cut(raw_data_30$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
raw_data_77$eigenbin <- cut(raw_data_77$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
raw_data_90$eigenbin <- cut(raw_data_90$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
raw_data_120$eigenbin <- cut(raw_data_120$eigencentrality, seq(0.0,1.1,0.04), right=FALSE)
```

```r
raw_data_30 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
raw_data_77 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
raw_data_90 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
raw_data_120 %>%
  ggplot(aes(x=eigencentrality, y=is_peak)) +
  geom_point()
```

```r
peak_data_30 = raw_data_30 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
peak_data_77 = raw_data_77 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
peak_data_90 = raw_data_90 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
peak_data_120 = raw_data_120 %>%
    group_by(eigenbin) %>%
    summarise(total_in_bin = n(), peak_counts = sum(is_peak == "True")) %>%
    mutate(percent = peak_counts/total_in_bin)
```

```r
peak_data_30 %>%
  filter(peak_counts < 10000) %>%
  ggplot(aes(x=eigenbin, y=peak_counts)) +
  geom_point()
```

```r
peak_data_77 %>%
  ggplot(aes(x=eigenbin, y=peak_counts)) +
  geom_point()
```

```r
peak_data_90 %>%
  ggplot(aes(x=eigenbin, y=peak_counts)) +
  geom_point()
```

```r
peak_data_120 %>%
  ggplot(aes(x=eigenbin, y=peak_counts)) +
  geom_point()
```

```r
peak_data_30 %>%
  ggplot(aes(x=eigenbin, y=percent)) +
  geom_point()
```

```r
peak_data_77 %>%
  ggplot(aes(x=eigenbin, y=percent)) +
  geom_point()
```

```r
peak_data_90 %>%
  ggplot(aes(x=eigenbin, y=percent)) +
  geom_point()
```

```r
peak_data_120 %>%
  ggplot(aes(x=eigenbin, y=percent)) +
  geom_point()
```
